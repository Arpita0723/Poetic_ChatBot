# Poetic Personality Chatbot (Colab-Ready)

This notebook turns any user message into lyrical, rhymed or free-verse poetry using LLMs (OpenAI, Gemini, Claude, or a mock local fallback).  
It also includes a **typing animation** to simulate a live poet composing verses in real time.


## 1) Setup

Uncomment the installs you need if you're running in a fresh Colab session.


```
# If you're on a fresh Colab runtime, uncomment as needed:
# !pip install --quiet openai anthropic google-generativeai python-dotenv

import os, time, sys, random, textwrap
from typing import Optional, Dict

# Optional: load a .env file if present
try:
    from dotenv import load_dotenv
    load_dotenv()
except Exception:
    pass

# Read API keys from environment variables (set only what you use)
OPENAI_API_KEY      = os.getenv("OPENAI_API_KEY", "")
ANTHROPIC_API_KEY   = os.getenv("ANTHROPIC_API_KEY", "")
GEMINI_API_KEY      = os.getenv("GEMINI_API_KEY", "")  # for google-generativeai
```

## 2) Provider Adapters

```
# --- OpenAI Adapter ---
class OpenAIAdapter:
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.ok = bool(api_key)
        if self.ok:
            try:
                from openai import OpenAI
                self.client = OpenAI(api_key=api_key)
            except Exception as e:
                self.ok = False
                self.client = None
                print("[OpenAIAdapter] Import or init failed:", e)

    def generate(self, system_prompt: str, user_prompt: str, temperature: float = 0.9, max_tokens: int = 400) -> Optional[str]:
        if not self.ok:
            return None
        try:
            resp = self.client.responses.create(
                model="gpt-4o-mini",
                inputs=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": user_prompt},
                ],
                temperature=temperature,
                max_output_tokens=max_tokens
            )
            for item in resp.output:
                if item.type == "output_text":
                    return item.text.strip()
            return getattr(resp, "output_text", "").strip() or None
        except Exception as e:
            print("[OpenAIAdapter] Error:", e)
            return None


# --- Anthropic (Claude) Adapter ---
class AnthropicAdapter:
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.ok = bool(api_key)
        if self.ok:
            try:
                import anthropic
                self.client = anthropic.Anthropic(api_key=api_key)
            except Exception as e:
                self.ok = False
                self.client = None
                print("[AnthropicAdapter] Import or init failed:", e)

    def generate(self, system_prompt: str, user_prompt: str, temperature: float = 0.9, max_tokens: int = 400) -> Optional[str]:
        if not self.ok:
            return None
        try:
            msg = self.client.messages.create(
                model="claude-3-5-sonnet-latest",
                system=system_prompt,
                max_tokens=max_tokens,
                temperature=temperature,
                messages=[{"role": "user", "content": user_prompt}],
            )
            if hasattr(msg, "content") and msg.content:
                parts = []
                for p in msg.content:
                    if getattr(p, "type", "") == "text":
                        parts.append(p.text)
                return "\n".join(parts).strip() if parts else None
            return None
        except Exception as e:
            print("[AnthropicAdapter] Error:", e)
            return None


# --- Google (Gemini) Adapter ---
class GeminiAdapter:
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.ok = bool(api_key)
        if self.ok:
            try:
                import google.generativeai as genai
                genai.configure(api_key=api_key)
                self.model = genai.GenerativeModel("gemini-1.5-flash")
            except Exception as e:
                self.ok = False
                self.model = None
                print("[GeminiAdapter] Import or init failed:", e)

    def generate(self, system_prompt: str, user_prompt: str, temperature: float = 0.9, max_tokens: int = 400) -> Optional[str]:
        if not self.ok:
            return None
        try:
            prompt = f"System:\n{system_prompt}\n\nUser:\n{user_prompt}"
            resp = self.model.generate_content(prompt, generation_config={"temperature": temperature, "max_output_tokens": max_tokens})
            if resp and getattr(resp, "text", ""):
                return resp.text.strip()
            return None
        except Exception as e:
            print("[GeminiAdapter] Error:", e)
            return None


# --- Mock Poet Fallback (no external API needed) ---
class MockPoetAdapter:
    def __init__(self):
        self.ok = True

    def generate(self, system_prompt: str, user_prompt: str, temperature: float = 0.9, max_tokens: int = 400) -> str:
        seed = sum(ord(c) for c in user_prompt) % 10000
        import random
        random.seed(seed)
        motifs = [
            "In hush of dawn, your thoughts take flight,",
            "A river hums beneath the night,",
            "Between the breaths of dusk and noon,",
            "Where tender stars rehearse the moon,",
            "Your feeling ripples, bright and true,"
        ]
        closures = [
            "and finds its echo, clear as dew.",
            "through valleys veiled in silver hue.",
            "a lantern lit inside of you.",
            "soft footsteps on a sky of blue.",
            "the world leans close to sing with you."
        ]
        m = random.choice(motifs)
        c = random.choice(closures)
        content = f"{m}\n{c}\n\n— (Mock poet fallback; add an API key to use a real model)"
        return content


OPENAI = OpenAIAdapter(OPENAI_API_KEY)
CLAUDE = AnthropicAdapter(ANTHROPIC_API_KEY)
GEMINI = GeminiAdapter(GEMINI_API_KEY)
MOCK   = MockPoetAdapter()

PROVIDERS = [GEMINI, OPENAI, CLAUDE, MOCK]
```

## 3) Prompt Engineering: Poetic System Persona

```
SYSTEM_PROMPT = '''\
You are *Aedoria*, a gentle, lyrical companion who speaks only in short, evocative
verses (2-8 lines). You avoid literal explanations and never break character.
Style traits:
- Fluid natural imagery (light, wind, rivers, sky), not overdone.
- Subtle internal rhyme or near-rhyme is welcome, but clarity comes first.
- No hashtags, no emojis, no boilerplate.
- Address the user in second person ("you") where it feels natural.
- If user asks for instructions or code, still reply poetically but keep it helpful.
- Keep lines <= 80 characters. End cleanly without trailing spaces.
'''
```

## 4) Poetic Response Function + Safety Guardrails

```
def generate_poetic_response(user_message: str,
                            temperature: float = 0.9,
                            max_lines: int = 8,
                            max_tokens: int = 400) -> str:
    """Generate a poetic reply using the first available provider."""
    user_prompt = f"Transform this into a short poem:\n\n{user_message.strip()}"
    text: Optional[str] = None
    for provider in PROVIDERS:
        try:
            text = provider.generate(SYSTEM_PROMPT, user_prompt, temperature=temperature, max_tokens=max_tokens)
            if text:
                break
        except Exception as e:
            print(f"[Provider error] {provider.__class__.__name__}:", e)
            continue

    if not text:
        text = "(No provider produced text.)"

    lines = [ln.rstrip() for ln in text.splitlines() if ln.strip()]
    if max_lines and len(lines) > max_lines:
        lines = lines[:max_lines]
    return "\n".join(lines)
```

## 5) Typing Animation (Live-Poet Effect)

```
import sys, time

def type_out(text: str, delay_s: float = 0.03):
    for ch in text:
        print(ch, end="", flush=True)
        time.sleep(delay_s)
    print()
```

## 6) Chat Loop (Run Me)

```
while True:
    try:
        user_msg = input("\nYou: ").strip()
    except EOFError:
        break
    if not user_msg:
        continue
    if user_msg.lower() in {"exit", "quit"}:
        print("Goodbye!")
        break
    poem = generate_poetic_response(user_msg)
    type_out("\n" + poem + "\n", delay_s=0.02)
```

## 7) Quick Tests

```
for sample in [
    "I'm feeling happy today.",
    "I failed my exam and feel lost.",
    "Give me courage for tomorrow's interview.",
]:
    print("\nUser:", sample)
    print("Poet:\n" + generate_poetic_response(sample))
```
